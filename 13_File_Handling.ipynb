**WITH OPEN**

    # This closes the file after performing the operations

```python
with open("file.txt") as f :
    print(f.read())
```

```python
f.closed
```

```python
with open ("file.txt","w") as f  : 
    f.write ("I am writable")
```

```python
with open("file.txt") as f :
    print(f.read())
```

**WRITING IN A FILE**

```python
# We are creating a new file here 
with open("new.txt","w") as f :
    f.write("Hey new file created")
```

```python
# Write in a existing file 
with open("new.txt","w") as f :
    f.write("This is the updated content")
```

```python
# Multiple lines 
with open("new.txt","w") as f :
    f.write("This is the updated content")
    f.write("This is a new line ")
```

**READING A DATA**

```python
with open("new.txt",'r') as f : 
    data = f.read()
    print(data)
```

```python
with open("new.txt",'r') as f : 
    data = f.read(4)
    print(data)

    # tell - tells the position of the pointer 
    data = f.tell()
    print(data)

    #seek - reposition the file handle pointer 
    f.seek(2)
    print(f.tell())

    print(f.read(1))
```

```python
with open("new.txt",'r') as f : 
    data = f.readlines()
    print(data)
```

```python
with open("new.txt",'r') as f : 
    data = f.readline()
    print(data)
```

```python
with open("new.txt",'r') as f : 
    data = f.readline()
    print(data)
```

**BINARY FILES**

```python
with open ("Sign.jpg","rb") as f : 
    data = f.read()
    print(data)
```

```python
with open ("Sign.jpg","rb") as f : 
    data = f.read()
    with open("new.jpeg","wb") as d : 
        d.write(data)
```
